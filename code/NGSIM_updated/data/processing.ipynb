```python
df = pd.read_csv('lankershim_trajectories_raw.txt', sep=' ', header=None)
df.columns = ['id', 'frame', 'n_frames_in_dataset', 'epoch', 'local_x', 'local_y', 'global_x', 'global_y', 'length', 'width', 'class', 'speed', 'acc', 'lane', 'origin', 'dest', 'intersection', 'section', 'direction', 'move', 'carind_front', 'carind_rear', 'dist_headway', 'time_headway']
```

```python
id_sorted  = np.sort(df.id.unique())
id_old2new = {t[1]: t[0]+1 for t in enumerate(id_sorted)}
```

After checking, there is an inconsistency in the data. Some car ids appear in mulitple frames with different lengths, widths...
According to https://data.transportation.gov/Automobiles/Next-Generation-Simulation-NGSIM-Vehicle-Trajector/8ect-6jqj, "repeats are not associated" so we need to give a new index to the repeated ids!

```python
n_dup_ids = df[dup==True]['id'].unique().size
print("there are %d duplicated ids" %n_dup_ids)
```

```python
dup_ids = df[dup==True]['id'].unique()
dup_idx = dup[dup==True].index
max_id  = df['id'].max()
dup2new_id = {ind: max_id+i+1 for (i,ind) in enumerate(dup_ids)}
```

```python
df.loc[dup==True, 'id'] = df.loc[dup==True]['id'].apply(lambda ind: dup2new_id[ind])
```

```python
df[dup==True].head()
```

```python
dup2 = df[['id','frame']].duplicated()
dup2[dup2==True].head()
```

Now every car appears at most once in each frame. Checking if lengths and widths are the same...

```python
lw = df[['id','length','width']].duplicated(keep=False)
assert(len(df[lw==True])==len(df)) 
```

Test passed! --> every id now corresponds to a unique vehicle
